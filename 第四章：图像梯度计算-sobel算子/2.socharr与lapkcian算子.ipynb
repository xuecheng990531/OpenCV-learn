```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
import imutils
```

```python
def show(img):
    plt.figure(dpi=100,figsize=(8,5))
    plt.xticks([]),plt.yticks([])
    plt.imshow(imutils.opencv2matplotlib(img))
    plt.show()
```

```python
circle=cv2.imread('/Users/lixuecheng/Downloads/OpenCV/代码+资料/第2-7章notebook课件/图像操作/lena.png')
show(circle)
```

```python
#sobel算子
kernal=np.ones((3,3),np.uint8)
sobel_x=cv2.Sobel(circle,cv2.CV_64F,1,0,3)
sobel_x=cv2.convertScaleAbs(sobel_x)
show(sobel_x)
```

```python
#sochar算子
schar_x=cv2.Scharr(circle,cv2.CV_64F,1,0)
schar_x=cv2.convertScaleAbs(schar_x)
show(schar_x)
```

```python
#拉普拉斯算子
laplasi=cv2.Laplacian(circle,cv2.CV_64F)
plt.imshow(laplasi)
```

```python
titles=["Orifin","sobel","Sochar","Laplasi"]
images=[circle,sobel_x,schar_x,laplasi]
for i in range(4):
    plt.subplot(2,2,i+1),plt.title(titles[i]),plt.imshow(images[i])
    plt.xticks([]),plt.yticks([])
plt.show()
```
