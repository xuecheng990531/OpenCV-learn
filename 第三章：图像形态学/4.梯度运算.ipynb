```python
#梯度运算可以得到图像的边缘
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
circle=cv.imread('/Users/lixuecheng/Downloads/OpenCV/代码+资料/第2-7章notebook课件/图像操作/pie.png')
plt.imshow(circle)
```

```python
#梯度运算
kernal=np.ones((3,3),np.uint8)
circle_gradian=cv.morphologyEx(circle,cv.MORPH_GRADIENT,kernal)
plt.imshow(circle_gradian)
```

```python
titles=['Origin','Gradient']
images=[circle,circle_gradian]
for i in range(2):
    plt.subplot(1,2,i+1),plt.title(titles[i]),plt.imshow(images[i])
    plt.xticks([]),plt.yticks([])
plt.show()
```
