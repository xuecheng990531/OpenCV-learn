```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
import imutils
```

```python
def show(img):
    plt.figure(dpi=100)
    plt.imshow(imutils.opencv2matplotlib(img))
    plt.show()
```

掌握图像的缩放平移旋转等操作，了解数字图像的仿射变换和透视变换 
API：cv.resize(src,dsize,fx,fy,interpolation=cv.INTER_LINER)
dsize:绝对尺寸，直接指定调整后的图像大小
fx/fy：相对尺寸，将dsize设置为none，然后将fx，fy设置为比例因子即可
interpolation：插值方法
    INTER_LINER：双线性插值法
    INTER_NEAREST：最近邻插值法
    INTER-AREA：像素区域重采样（默认方法）
    INTER_CUBIC：双三次插值

```python
img=cv.imread('/Users/lixuecheng/Downloads/OpenCV/代码+资料/第2-7章notebook课件/图像操作/cat.jpg')
```

```python
#获取图像的尺寸信息,绝对尺寸为例子
rows,cols=img.shape[:2]
res=cv.resize(img,(2*rows,2*cols),interpolation=cv.INTER_LINEAR)
plt.imshow(imutils.opencv2matplotlib(res))
```

```python
#相对尺寸设置
res1=cv.resize(img,None,fx=0.5,fy=0.5)
```

```python
#图像显示
plt.figure(dpi=120,figsize=(10,8))
plt.subplot(131),plt.title("Orign"),plt.imshow(imutils.opencv2matplotlib(img))
plt.subplot(132),plt.title("Absolute"),plt.imshow(imutils.opencv2matplotlib(res))
plt.subplot(133),plt.title("Relative"),plt.imshow(imutils.opencv2matplotlib(res1))
plt.show()
```
