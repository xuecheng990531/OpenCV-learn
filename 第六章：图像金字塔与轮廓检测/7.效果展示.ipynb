```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
import imutils
```

```python
def show(img):
    plt.xticks([]),plt.yticks([])
    plt.imshow(imutils.opencv2matplotlib(img))
    plt.show()
```

```python
#如何匹配多个符合模版的位置呢？
src_orign=cv.imread('/Users/lixuecheng/Downloads/OpenCV/代码+资料/第2-7章notebook课件/图像操作/lena.png')
src=cv.imread('/Users/lixuecheng/Downloads/OpenCV/代码+资料/第2-7章notebook课件/图像操作/lena.png',0)

template_orign=cv.imread('/Users/lixuecheng/Downloads/OpenCV/代码+资料/第2-7章notebook课件/图像操作/lena头像.jpg')
template=cv.imread('/Users/lixuecheng/Downloads/OpenCV/代码+资料/第2-7章notebook课件/图像操作/lena头像.jpg',0)
```

```python
#获得模版的长和宽,用来画处矩形信息
h,w=template.shape[:2]
```

```python
res=cv.matchTemplate(src,template,5)
```

```python
minval,maxval,minloc,maxloc=cv.minMaxLoc(res)
```

```python
topleft=maxloc
```

```python
bottomright=(topleft[0]+w,topleft[1]+h)
```

```python
cv.rectangle(src_orign,topleft,bottomright,(0,255,255),2)
```

```python
plt.imshow(imutils.opencv2matplotlib(src_orign))
plt.show()
```
