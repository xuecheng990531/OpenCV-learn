```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
import imutils
```

```python
def show(img):
    plt.figure(dpi=200)
    plt.imshow(imutils.opencv2matplotlib(img))
    plt.xticks([]),plt.yticks([])
    plt.show()
```

```python
im=cv.imread('/Users/lixuecheng/Downloads/OpenCV/代码+资料/第2-7章notebook课件/图像操作/AM.png')
print(im.shape)
show(im)
```

```python
#向上操作pyrUP(Src)
up_img=cv.pyrUp(im)
print(up_img.shape)
show(up_img)
```

```python
#下操作pyrDown(Src)
down_img=cv.pyrDown(im)
print(down_img.shape)
show(down_img)
```
