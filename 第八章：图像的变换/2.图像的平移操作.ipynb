```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
import imutils
```

```python
def show(img):
    plt.imshow(imutils.opencv2matplotlib(img))
```

# API:cv.wrapAffine(img,M,dsize)<br>img:输入的图像<br>M:2*3的移动矩<br>dsize:输出图像的大小(width,height)

```python
#需求：将图像的像素点移动（50，100）的距离：
img=cv.imread('/Users/lixuecheng/Downloads/OpenCV/代码+资料/第2-7章notebook课件/图像操作/cat.jpg')
rows,cols=img.shape[:2]
```

```python
M=np.float32([[1,0,100],[0,1,50]])
dst=cv.warpAffine(img,M,(cols,rows))
```

```python
plt.figure(dpi=120)
plt.subplot(121),show(img)
plt.subplot(122),show(dst)
plt.show
```
