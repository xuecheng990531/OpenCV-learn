```python
import matplotlib.pyplot as plt
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
import imutils
```

```python
def show(img):
    plt.figure(dpi=140)
    plt.xlim(0,255)
    plt.plot(img)
```

```python
img=cv.imread('/Users/lixuecheng/Downloads/OpenCV/代码+资料/第2-7章notebook课件/图像操作/clahe.jpg',0)
plt.imshow(imutils.opencv2matplotlib(img))
```

```python
#绘制直方图
hinst=cv.calcHist([img],[0],None,[256],[0,256])
plt.hist(hinst.ravel(),256)
plt.show()
```

```python
#均衡优化
new_hist=cv.equalizeHist(img)
plt.hist(new_hist.ravel(),255)
plt.show()
```

```python
#对原始图像加上掩码mask
mask=np.zeros(img.shape[:2],np.uint8)
mask[100:250,100:350]=255
plt.imshow(imutils.opencv2matplotlib(mask))
```

```python
mask_add=cv.bitwise_and(img,img,mask=mask)
plt.imshow(imutils.opencv2matplotlib(mask_add))
```

```python
#加入掩码后的效果
hinst_masked=cv.calcHist([img],[0],mask_add,[256],[0,256])
plt.hist(hinst_masked.ravel(),256)
plt.show()
```

```python
#使用自定义均衡器的效果
result=cv.createCLAHE(clipLimit=2.0,tileGridSize=(8,8))
result_img=result.apply(img)
plt.imshow(imutils.opencv2matplotlib(result_img))
```

```python
hist_result=cv.calcHist([result_img],[0],None,[256],[0,256])
plt.hist(hist_result.ravel(),256)
plt.show()
```
