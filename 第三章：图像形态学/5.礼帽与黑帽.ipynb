```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
import imutils
```

```python
def show(img):
    plt.figure(dpi=100,figsize=(7,6))
    plt.xticks([]),plt.yticks([])
    plt.imshow(imutils.opencv2matplotlib(img))
    plt.show()
```

```python
dige=cv.imread('/Users/lixuecheng/Downloads/OpenCV/代码+资料/第2-7章notebook课件/图像操作/dige.png')
```

```python
#礼帽运算：初始图像-开运算的结果
kernal=np.ones((3,3),np.uint8)
dige_tophat=cv.morphologyEx(dige,cv.MORPH_TOPHAT,kernal)
show(dige_tophat)
```

```python
#黑帽运算
dige_blackhat=cv.morphologyEx(dige,cv.MORPH_BLACKHAT,kernal)
show(dige_blackhat)
```

```python
titles=['Origin','Tophat','Blackhat']
images=[dige,dige_tophat,dige_blackhat]
for i in range(3):
    plt.subplot(1,3,i+1),plt.title(titles[i]),plt.imshow(images[i])
plt.show()
```
