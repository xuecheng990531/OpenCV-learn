```python
import cv2
import matplotlib.pyplot as plt
import imutils
```

```python
def plt_img_show(img):
    plt.figure(dpi=100,figsize=(4,4))
    plt.imshow(imutils.opencv2matplotlib(img))
    plt.xticks([]),plt.yticks([])
```

```python
img=cv2.imread('/Users/lixuecheng/Downloads/OpenCV/代码+资料/第2-7章notebook课件/图像操作/lenaNoise.png')
plt_img_show(img)
```

```python
#均值滤波blur()
img_blur=cv2.blur(img,(3,3))
plt_img_show(img_blur)#均值滤波过后噪音稍微的减少了
```

```python
#方框滤波,normalize=false，当一旦超过255，就直接取255，因此图像白化
img_box=cv2.boxFilter(img,-1,(3,3),normalize=False)
plt_img_show(img_box)
```

```python
#方框滤波,normalize=true.相当于均值滤波
img_box=cv2.boxFilter(img,-1,(3,3),normalize=True)
plt_img_show(img_box)
```

```python
#高斯滤波
img_gaosi=cv2.GaussianBlur(img,(3,3),0.8)
plt_img_show(img_gaosi)
```

```python
#中值滤波:按照设定的范围从小到大排列找到中间的值，作为处理结果
img_medvalue=cv2.medianBlur(img,3)
plt_img_show(img_medvalue)
```

```python
#如何展示所有结果
import numpy as np
res=np.hstack((img,img_box,img_blur,img_gaosi,img_medvalue))
cv2.imshow("all",res)
cv2.waitKey(10)
cv2.destroyAllWindows()
```
