```python
import cv2 as cv
import matplotlib.pyplot as plt
import imutils
def show(img):
    plt.figure(dpi=150)
    plt.imshow(imutils.opencv2matplotlib(img))
    plt.xticks([]),plt.yticks([])
    plt.show()
```

```python
#转换成二值图像
img=img=cv.imread('/Users/lixuecheng/Downloads/OpenCV/代码+资料/第2-7章notebook课件/图像操作/contours2.png')
gray=cv.cvtColor(img,cv.COLOR_BGR2GRAY)
res,img_final=cv.threshold(gray,127,255,cv.THRESH_BINARY)
```

```python
show(img_final)
```

```python
contours,layer=cv.findContours(img_final,cv.RETR_TREE,cv.CHAIN_APPROX_NONE)
```

```python
#绘画
img_copy=img_final.copy()
res=cv.drawContours(img,contours,-1,(0,0,255),1)
show(res)
```

```python
#轮廓近似
cnt=contours[0]
ep=0.1*cv.arcLength(cnt,True)
approx=cv.approxPolyDP(cnt,ep,True)

drwa=img_final.copy()
approx_image=cv.drawContours(img,[approx],-1,(0,0,255),2)
show(approx_image)
```
