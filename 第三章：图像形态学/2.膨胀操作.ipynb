```python
import cv2 as cv
import imutils
import matplotlib.pyplot as plt
```

```python
def imshow(img):
    plt.xticks([]),plt.yticks([])
    plt.imshow(img)
```

```python
#载入原始迪哥图像
dige=cv.imread('/Users/lixuecheng/Downloads/OpenCV/代码+资料/第2-7章notebook课件/图像操作/dige.png')
imshow(dige)
```

```python
import numpy as np
kernal=np.ones((3,3),np.uint8)
#腐蚀迪哥的操作
dige_erod=cv.erode(dige,kernal,iterations=1)
imshow(dige_erod)
```

```python
#膨胀迪哥
img_pengzhang=cv.dilate(dige_erod,kernal,iterations=1)
imshow(img_pengzhang)
```
