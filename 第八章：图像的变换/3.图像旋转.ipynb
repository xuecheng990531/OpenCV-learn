```python
import cv2 as cv
import matplotlib.pyplot as plt
import imutils
import numpy as np
```

```python
img=cv.imread('/Users/lixuecheng/Downloads/OpenCV/代码+资料/第2-7章notebook课件/图像操作/cat.jpg')
rows,cols=img.shape[:2]
```

```python
#生成旋转矩阵
res=cv.getRotationMatrix2D((cols/2,rows/2),90,1)
```

```python
#进行旋转变化
dst=cv.warpAffine(img,res,(cols,rows))
```

```python
plt.figure(dpi=200)
plt.subplot(121),plt.title("Orign"),plt.imshow(imutils.opencv2matplotlib(img))
plt.subplot(122),plt.title("Rotate"),plt.imshow(imutils.opencv2matplotlib(dst))
plt.show()
```
