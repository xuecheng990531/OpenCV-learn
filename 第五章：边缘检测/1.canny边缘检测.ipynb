```python
import cv2 as cv
import matplotlib.pyplot as plt
import imutils
```

```python
car=cv.imread('/Users/lixuecheng/Downloads/OpenCV/代码+资料/第2-7章notebook课件/图像操作/car.png',cv.IMREAD_GRAYSCALE)
plt.imshow(imutils.opencv2matplotlib(car))
```

```python
min_canny=cv.Canny(car,50,100)
max_canny=cv.Canny(car,100,250)
titles=["min_canny","max_canny"]
images=[min_canny,max_canny]
for i in range(2):
    plt.figure(figsize=(8,8),dpi=150)
    plt.subplot(2,1,i+1),plt.title(titles[i]),plt.imshow(imutils.opencv2matplotlib(images[i]))
plt.show()
```
