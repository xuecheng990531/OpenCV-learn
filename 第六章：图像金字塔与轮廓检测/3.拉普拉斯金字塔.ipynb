```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
import imutils
```

```python
def show(img):
    plt.figure(dpi=200)
    plt.imshow(imutils.opencv2matplotlib(img))
    plt.xticks([]),plt.yticks([])
    plt.show()
```

```python
im=cv.imread('/Users/lixuecheng/Downloads/OpenCV/代码+资料/第2-7章notebook课件/图像操作/AM.png')
```

```python
lap_image=im-cv.pyrUp(cv.pyrDown(im))
lap_image.shape
```

```python
show(lap_image)
```
