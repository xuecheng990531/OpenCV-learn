```python
import cv2 as cv
```

```python
img=cv.imread('/Users/lixuecheng/Downloads/OpenCV/代码+资料/第2-7章notebook课件/图像操作/cat.jpg')
```

```python
top_size,bottom_size,left_size,right_size=(50,50,50,50)#把上下左右的边距设置为50
```

```python
#四种填充方式
replacate=cv.copyMakeBorder(img,top_size,bottom_size,left_size,right_size,borderType=cv.BORDER_REPLICATE)
reflect = cv.copyMakeBorder(img, top_size, bottom_size, left_size, right_size,cv.BORDER_REFLECT)
reflect101 = cv.copyMakeBorder(img, top_size, bottom_size, left_size, right_size, cv.BORDER_REFLECT_101)
wrap = cv.copyMakeBorder(img, top_size, bottom_size, left_size, right_size, cv.BORDER_WRAP)
constant = cv.copyMakeBorder(img, top_size, bottom_size, left_size, right_size,cv.BORDER_CONSTANT, value=0)

title=["Origin","Reflect","Reflect101","Wrap","Constant","Replacate"]
images=[img,replacate,reflect,reflect101,wrap,constant]
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6),dpi=120)

for i in range(6):
    plt.title(title[i]),plt.subplot(2,3,i+1),plt.imshow(images[i])
    plt.xticks([]),plt.yticks([])

plt.show()
```
