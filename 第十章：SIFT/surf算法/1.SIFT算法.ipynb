```python
import cv2 as cv
import matplotlib.pyplot as plt
import imutils
import numpy as np
```

```python
img=cv.imread('/Users/lixuecheng/Desktop/python/OpenCV/image/央视总部大楼.jpeg')
img2=cv.imread('/Users/lixuecheng/Desktop/python/OpenCV/image/央视总部大楼.jpeg')
gray=cv.cvtColor(img,cv.COLOR_BGR2GRAY)
```

```python
#实例化sift
sift=cv.xfeatures2d.SIFT_create()
```

```python
'''
利用sift.detectandcompute()检测关键点并计算,
gray:进行关键点检测的图像，必须是灰度图
kp:关键点信息，包括尺度，位置，方向等信息
des:关键点描述符，每个关键点对应128个梯度信息的特征向量
'''
kp,des=sift.detectAndCompute(gray,None)
```

```python
#将关键点信息结果绘制在图像上
cv.drawKeypoints(img,kp,img2,(0,255,255),flags=cv.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
'''
outputimg也可以是原始的图像
color（b,g,r)
flags绘图功能设置
'''
```

```python
plt.figure(dpi=250)
plt.subplot(121),plt.title("Orign"),plt.imshow(imutils.opencv2matplotlib(img))
plt.subplot(122),plt.title("After"),plt.imshow(imutils.opencv2matplotlib(img2))
```
