```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
kernal=np.ones((2,2),np.uint8)
kernal
```

```python
img=cv.imread('/Users/lixuecheng/Downloads/OpenCV/代码+资料/第2-7章notebook课件/图像操作/dige.png')
plt.imshow(img)
```

```python
#iterations=1表示迭代次数为1次
erode_img=cv.erode(img,kernal,iterations=1)
plt.imshow(erode_img)
plt.xticks([]),plt.yticks([])
```

```python
img_circle=cv.imread('/Users/lixuecheng/Downloads/OpenCV/代码+资料/第2-7章notebook课件/图像操作/pie.png')
kernal1=np.ones((30,30),np.uint8)
erode1=cv.erode(img_circle,kernal1,iterations=1)
erode2=cv.erode(img_circle,kernal1,iterations=2)
erode3=cv.erode(img_circle,kernal1,iterations=3)
res=np.hstack((img_circle,erode1,erode2,erode3))

plt.figure(dpi=150,figsize=(10,8))
plt.xticks([]),plt.yticks([])
plt.imshow(res)#因为腐蚀处理的图像一般都是二值图像，不用加imutils包
```
