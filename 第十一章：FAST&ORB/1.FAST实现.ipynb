```python
import cv2 as cv
import matplotlib.pyplot as plt
import imutils
import numpy as np
```

```python
#灰度图
img2=cv.imread('/Users/lixuecheng/Desktop/python/OpenCV/image/央视总部大楼.jpeg')
img3=cv.imread('/Users/lixuecheng/Desktop/python/OpenCV/image/央视总部大楼.jpeg')
img=cv.imread('/Users/lixuecheng/Desktop/python/OpenCV/image/央视总部大楼.jpeg')
grayimg=cv.cvtColor(img,cv.COLOR_BGR2GRAY)
```

```python
#默认开启非极大值抑制，所以不用写nonemaxsuppression
fast1=cv.FastFeatureDetector_create(threshold=20)
```

```python
kp=fast1.detect(grayimg,None)
'''
grayimg:灰度图像
kp：关键点信息
'''
```

```python
#在图像上标记处监测点
cv.drawKeypoints(img,kp,img2,(0,0,255))
```

```python
fast2=cv.FastFeatureDetector_create(threshold=20,nonmaxSuppression=False)
```

```python
kp2=fast2.detect(grayimg,None)
```

```python
dst=cv.drawKeypoints(img,kp2,img3,(0,255,255))
```

```python
plt.figure(dpi=300)
plt.subplot(121),plt.imshow(imutils.opencv2matplotlib(img3)),plt.title("True")
plt.subplot(122),plt.imshow(imutils.opencv2matplotlib(img2)),plt.title("False")
```
