```python
import matplotlib.pyplot as plt
import cv2 as cv
```

```python
img=cv.imread('/Users/lixuecheng/Downloads/OpenCV/代码+资料/第2-7章notebook课件/图像操作/cat.jpg',0)

'''
cv.calcHist([img],[0],None,[256],[0,256])
[img]代表原图像
[0]表示色彩通道
None表示对整个区域进行绘制直方图
[256]代表BIN的数目
[0,256]代表像素值的范围
'''
hinst=cv.calcHist([img],[0],None,[256],[0,256])
plt.figure(dpi=100)
#plt.hist()：直方图，一种特殊的柱状图。
plt.hist(img.ravel(),256)
plt.show()
```

```python
#显示多色值要使用原来的彩色图像
img_orign=cv.imread('/Users/lixuecheng/Downloads/OpenCV/代码+资料/第2-7章notebook课件/图像操作/cat.jpg')
color=('b','g','r')

#enumerate() 函数用于将一个可遍历的数据对象
for i,col in enumerate(color):
    hinst=cv.calcHist([img_orign],[i],None,[255],[0,256])
    plt.plot(hinst,color=col)
    plt.xlim([0,256])#限制x轴的范围为0-256
```
