```python
import cv2 as cv
import matplotlib.pyplot as plt
import imutils
import numpy as np
```

```python
img=cv.imread('/Users/lixuecheng/Desktop/python/OpenCV/image/央视总部大楼.jpeg')
img2=cv.imread('/Users/lixuecheng/Desktop/python/OpenCV/image/央视总部大楼.jpeg')
```

```python
#转换为灰度图
gray=cv.cvtColor(img,cv.COLOR_BGR2GRAY)
```

```python
#角点检测,需要1000个角点，最低质量0.01，角点之间的距离10
conner=cv.goodFeaturesToTrack(gray,1000,0.01,10)
```

```python
#绘制
for i in conner:
    '''
    参数：
    image:它是要在其上绘制圆的图像。
    center_coordinates：它是圆的中心坐标。坐标表示为两个值的元组，即(X坐标值，Y坐标值)。
    radius:它是圆的半径。
    color:它是要绘制的圆的边界线的颜色。对于BGR，我们通过一个元组。例如：(255，0，0)为蓝色。
    thickness:它是圆边界线的粗细像素。厚度-1像素将以指定的颜色填充矩形形状。

    The problem is just parsing the (x, y) or the (pts1[0][0], pts1[0][1]) to integers you just have to do the following (int(pts1[0][0]), int(pts1[0][1])) in the circle function->x,y可能不是整数，因此要进行类型的转换
    '''
    x,y=i.ravel()
    cv.circle(img2,(int(x),int(y)),3,(0,0,255),-1)
```

```python
plt.figure(dpi=200)
plt.subplot(121),plt.title("orign"),plt.imshow(imutils.opencv2matplotlib(img))
plt.subplot(122),plt.title("after"),plt.imshow(imutils.opencv2matplotlib(img2))
```
