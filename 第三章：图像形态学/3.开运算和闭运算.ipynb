```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
```

```python
dige=cv.imread('/Users/lixuecheng/Downloads/OpenCV/代码+资料/第2-7章notebook课件/图像操作/dige.png')
```

```python
#开运算=先腐蚀再膨胀
kernal=np.ones((3,3),np.uint8)
dige_open=cv.morphologyEx(dige,cv.MORPH_OPEN,kernal)
plt.imshow(dige_open)
```

```python
#闭运算=先膨胀，再腐蚀
dige_close=cv.morphologyEx(dige,cv.MORPH_CLOSE,kernal)
plt.imshow(dige_close)
```
