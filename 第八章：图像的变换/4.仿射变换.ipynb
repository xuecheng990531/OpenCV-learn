```python
import cv2 as cv
import matplotlib.pyplot as plt
import imutils 
import numpy as np
```

```python
img=cv.imread('/Users/lixuecheng/Downloads/OpenCV/代码+资料/第2-7章notebook课件/图像操作/cat.jpg')
rows,cols=img.shape[:2]
```

```python
#创建变换矩阵,指定三个点，接着指定仿射变换后的三个点
pts1=np.float32([[50,50],[200,50],[50,200]])
pts2=np.float32([[100,100],[200,50],[200,250]])
M=cv.getAffineTransform(pts1,pts2)
```

```python
#完成仿射变换
dst=cv.warpAffine(img,M,(cols,rows))
```

```python
plt.figure(dpi=200)
plt.subplot(121),plt.title("Orign"),plt.imshow(imutils.opencv2matplotlib(img))
plt.subplot(122),plt.title("after"),plt.imshow(imutils.opencv2matplotlib(dst))
```
