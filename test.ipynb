```python
import cv2 as cv
import matplotlib.pyplot as plt
import imutils
import numpy as np
```

```python
cap=cv.VideoCapture('/Users/lixuecheng/Desktop/2021-7-28 15.45拍摄的影片.mov')
```

```python
ret,frame=cap.read()
```

```python
r,h,c,w=400,200,600,300
window=(c,r,h,w)
```

```python
roi=frame[r:r+h,c:c+w]
```

```python
hsv_roi=cv.cvtColor(roi,cv.COLOR_BGR2HSV)
roi_hist=cv.calcHist(hsv_roi,[0],None,[180],[0,180])
```

```python
cv.normalize(roi_hist,roi_hist,0,255,cv.NORM_MINMAX)
```

```python
term=(cv.TERM_CRITERIA_EPS|cv.TERM_CRITERIA_COUNT,10,1)
```

```python
while (True):
    ret,frame=cap.read()
    if ret==True:
        hsv=cv.cvtColor(frame,cv.COLOR_BGR2HSV)
        dst=cv.calcBackProject([hsv],[0],roi_hist,[0,180],1)
        ret,windowtrack=cv.meanShift(dst,window,term)

        x,y,w,h=windowtrack

        img2=cv.rectangle(frame,(x,y),(x+w,y+h),(0,255,255),2)
        cv.imshow("frame",img2)

        if cv.waitKey(60)&0xFF==27:
            break
cap.release()
cv.destroyAllWindows()
```
