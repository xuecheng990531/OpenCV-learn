```python
import cv2 as cv
import matplotlib.pyplot as plt
import imutils
import numpy as np
```

```python
img=cv.imread('/Users/lixuecheng/Desktop/python/OpenCV/image/角点.jpeg')
gray=cv.cvtColor(img,cv.COLOR_BGR2GRAY)
img32=np.float32(gray)
```

```python
harris=cv.cornerHarris(img32,2,3,0.04)
```

```python
#将大于0.06harrismax的地方标记成红色
img[harris>0.06*harris.max()]=[0,0,255]
```

```python
plt.figure(dpi=200)
plt.imshow(imutils.opencv2matplotlib(img))
```
