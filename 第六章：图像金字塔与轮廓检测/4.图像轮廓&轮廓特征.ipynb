```python
import cv2 as cv
import matplotlib.pyplot as plt
import imutils
```

```python
def show(img):
    plt.figure(dpi=150)
    plt.imshow(imutils.opencv2matplotlib(img))
    plt.xticks([]),plt.yticks([])
    plt.show()
```

```python
#为了方便检测，采用二值图像进行分析
img=cv.imread('/Users/lixuecheng/Downloads/OpenCV/代码+资料/第2-7章notebook课件/图像操作/contours.png')
gray=cv.cvtColor(img,cv.COLOR_BGR2GRAY)
ret,thresh=cv.threshold(gray,127,255,cv.THRESH_BINARY)
show(thresh)
```

```python
#binary得到的结果就是二值之后的结果，contours是一些轮廓点,hierarchy代表的是层级
contours,hierarchy=cv.findContours(thresh,cv.RETR_TREE,cv.CHAIN_APPROX_NONE)
```

```python
#绘制轮廓
draw_img=thresh.copy()
#-1代表画出所有轮廓，（0，0，255）代表BGR，2代表线条宽度,ig就是原图，contours代表轮廓是什么
res=cv.drawContours(img,contours,-1,(255,0,0),2)
show(res)
```

```python
titles=['Origin','result']
images=[thresh,res]
for i in range(2):
    plt.subplot(1,2,i+1),plt.title(titles[i]),plt.imshow(imutils.opencv2matplotlib(images[i]))
    plt.xticks([]),plt.yticks([])
plt.show()
```

```python
#面积函数countarea(),事先需要把contours[0]索引出来
cnt=contours[0]
area=cv.contourArea(cnt)
area
```

```python
#计算周长archlength(src，true/false)，true代表是一个闭合的轮廓，flase是闭合的轮廓
cnt_lenght=cv.arcLength(cnt,True)
cnt_lenght
```
